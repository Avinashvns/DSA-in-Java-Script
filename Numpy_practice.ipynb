User Input Creation Array -

```python
import numpy as np
n=int(input("Enter number of elements :"))
arr=[]

for i in range(n):
  val=int(input(f"Enter element {i+1}: "))
  arr.append(val)

a=np.array(arr)

print("Array is :" , a)
```

```python
import numpy as np
a=np.array([10,20,30,40,50])

# Shape
print(a.shape)

# array dimension
print(a.ndim)

print(type(a))

print(a.dtype)

# all element
print(a.size)

# 1 element Size in memory
print(a.itemsize)

# Total Memory
print(a.nbytes)
```
